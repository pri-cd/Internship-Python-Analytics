#### **Exercises**
- **`header = 0`** Tells the `pd.read_html` to use the first row as the header row.







```python
df_titanic_html = pd.read_html('https://github.com/pri-cd/Internship-Python-Analytics/raw/main/Datasets/Titanic.html', header=0)
df_titanic_html[0].head()
```
