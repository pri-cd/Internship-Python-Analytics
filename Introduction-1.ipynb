## Explanation of X and Y in a Supervised Linear Regression Model

In a supervised learning model, such as **Linear Regression**, we aim to predict a dependent variable (often called the "target") based on one or more independent variables (often called "features").

### X (Independent Variable)
- **X** represents the input feature(s) of the model, also known as the **independent variable**.
- In this case, **X** is the **years of experience**.
- It is called "independent" because its values are not dependent on any other variable in the dataset.

### Y (Dependent Variable)
- **Y** represents the output of the model, also known as the **dependent variable** or the **target**.
- In this case, **Y** is the **salary**.
- It is called "dependent" because its values are dependent on the input (X), and our goal is to predict this value based on the provided features.

### In Our Example:
- **X (Experience)**: The number of years of experience someone has.
- **Y (Salary)**: The salary that is earned based on those years of experience.

In a linear regression model, we use the relationship between X and Y to create a straight line that best fits the data, allowing us to predict the salary (Y) for a given number of years of experience (X).


````python
print(f"Null, NA Elements: {df_salary['Experience Years'].isna().sum()}, {df_salary['Experience Years'].isnull().sum()}")
````

- What we have to **Predict** ? `(Y)`
- What we have to use to **Predict** Y ? `(X)`

#### **Single Square Brackets** `[]`
- **Usage**: To select a single column from a DataFrame and get it as a **pandas Series**.
- **Why**: The target variable `y` is typically used as a 1D array for machine learning models.


#### **Double Square Brackets** `[[]]`
- **Usage**: To select a single column from a DataFrame and get it as a **pandas DataFrame**.
- **Why**: The feature matrix `X` should be a 2D array, even if there's only one feature. Machine learning models **expect** the input features to be in this format.


````python
y = df_salary['Salary']
x = df_salary[['Experience Years']]

print(f"Shape of Y: {y.shape} & X: {x.shape}")
````

##  Quick Reference: Shape `(40, 1)` vs `(40,)` in Linear Regression

In linear regression, understanding the shape of **X** (independent variable) and **y** (dependent variable) is crucial!

### X (Independent Variable) 
- **Shape (40, 1)**:
  - This is a 2D array with 40 rows and 1 column.
  - **Why?**  Machine learning models like linear regression expect input features in a 2D format.
  -  **Example**:
    ```python
    X = df[['Experience Years']]  # 2D array with 40 rows, 1 column
    ```

### y (Dependent Variable) 
- **Shape (40,)**:
  - This is a 1D array with 40 elements.
  - **Why?**  It's fine for `y` (target) to be a 1D array because you're predicting just one value per row.
  -  **Example**:
    ```python
    y = df['Salary']  # 1D array with 40 values
    ```

###  Summary:
- **Use (40, 1)** for `X` (features) 
- **Use (40,)** for `y` (target) 

Remember:
- **X** should be 2D (even for one feature).
- **y** can stay 1D.


##  Train-Test Split

**Train-Test Split** is a technique used to evaluate the performance of a machine learning model. It involves dividing your dataset into two subsets: one for training the model and one for testing its performance.

### **Why Use Train-Test Split?**
- **Train the Model**: Use a portion of the data to train the model.
- **Evaluate the Model**: Use the remaining data to test how well the model performs on unseen data.


````python
from sklearn.model_selection import train_test_split

# Why "random_state"?
# Because, We don't want to randomise our Train, Test Split, Everytime We Run It.
# Also, Always be ready to present & With Confidence!.
XTr, XTe, YTr, YTe = train_test_split(x, y, train_size=0.75, test_size=0.25, random_state=42)

print(f"Shape of XTr: {XTr.shape} & Xte: {XTe.shape}")
print(f"Shape of YTr: {YTr.shape} & YTe: {YTe.shape}")

XTr.head()
````

### **Starting Analysis**

````python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score


# Created A Simple Instance of Linear Regression!- & Fitting!
Model = LinearRegression()
Model.fit(XTr, YTr)

# Start Prediction!
yPred = Model.predict(XTe)
````

````python
df_results = pd.DataFrame({
    'Actual': YTe,
    'Predicted': yPred
})

df_results.head()
````

````python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.scatterplot(x=df_results['Actual'], y=df_results['Predicted'], color='blue', label='Predicted')
plt.plot([df_results['Actual'].min(), df_results['Actual'].max()],
         [df_results['Actual'].min(), df_results['Actual'].max()], color='red', label='Ideal Prediction')

plt.title('Actual vs Predicted Salary')
plt.xlabel('Actual Salary')
plt.ylabel('Predicted Salary')
plt.legend()
plt.grid(True)
plt.show()
````

### **Checking Accuracy**
- We've done The Model Fitting & Analysis in above Cell.
- Now, We've to check the accuracy and MSE.

````python
# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(YTe, yPred)

# Calculate R-squared score
r2 = r2_score(YTe, yPred)
print(f"R-squared (R) Score: {round((r2 * 100), 4)} % Accuracy!")
print(f"Mean Squared Error: {mse}")
````

---
- We got about **94%** Accuracy!-
- With, Very low **MSE** (**Mean Squared Error**).
