## **Importing Important Libraries**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://github.com/pri-cd/Internship-Python-Analytics/raw/main/Datasets/Diabetes.csv')
df.head()
```

```python
print(f"The Shape of DataFrame: {df.shape}")
print(f"The Columns are: {df.columns.tolist()}")
print(end='\n\n')

# Checking for Null Values:
for column in df.columns:
  print(f"Null Values in Column: {column}: {df[column].isnull().sum()}")
```

```python
df.describe()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df[['pregnancies', 'glucose', 'diastolic', 'triceps', 'insulin', 'bmi', 'dpf']] = scaler.fit_transform(df[['pregnancies', 'glucose', 'diastolic', 'triceps', 'insulin', 'bmi', 'dpf']])
df.head()
```

#### **Defining The `X` & `Y`.**

```python
x = df[['pregnancies', 'glucose', 'diastolic', 'triceps', 'insulin', 'bmi', 'dpf', 'age']]
y = df['diabetes']

print(f"Shape of X:{x.shape} & Y:{y.shape}")
```

#### **Splitting Dataset**
- We Split in **70-30** Ratio

```python
from sklearn.model_selection import train_test_split
XTr, XTe, YTr, YTe = train_test_split(x, y, train_size=0.75, test_size=0.25, random_state=100)
```

```python
from sklearn.linear_model import LogisticRegression
LRModel = LogisticRegression(max_iter=500, C=5)
LRModel.fit(XTr, YTr)
```

```python
LRYPr = LRModel.predict(XTe)
```

```python
from sklearn.metrics import accuracy_score

LRaccuracy = accuracy_score(YTe, LRYPr)
print(f'Accuracy: {LRaccuracy}')
```

#### **Retrying With KNN**

```python
from sklearn.neighbors import KNeighborsClassifier
KNNModel = KNeighborsClassifier()
KNNModel.fit(XTr, YTr)
```

```python
KNNYPr = KNNModel.predict(XTe)
```

```python
KNNaccuracy = accuracy_score(YTe, KNNYPr)
print(f'Accuracy: {KNNaccuracy}')
```

#### **Another One**

```python
from sklearn.tree import DecisionTreeClassifier
DTModel = DecisionTreeClassifier()
DTModel.fit(XTr, YTr)
DTPr = DTModel.predict(XTe)
```

```python
DTAccuracy = accuracy_score(YTe, DTPr)
print(f'Accuracy: {DTAccuracy}')
```
